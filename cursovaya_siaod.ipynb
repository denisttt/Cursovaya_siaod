```python
import csv 
from collections import defaultdict 
 
filecsv = 'data.csv' 
 
def read_sales_data(file_path): 
    with open('data.csv','r', encoding='utf-8') as file: 
        csv_reader = csv.DictReader(file) 
        sales_data = [] 
        for row in csv_reader: 
            sales_data.append(row) 
    return sales_data 
 
def print_sales_data(data): 
    for sale in data: 
        print(f"Номер заказа: {sale['Номер заказа']}") 
        print(f"Дата заказа: {sale['Дата заказа']}") 
        print(f"Название товара: {sale['Название товара']}") 
        print(f"Категория товара: {sale['Категория товара']}") 
        print(f"Количество продаж: {sale['Количество продаж']}") 
        print(f"Цена за единицу: {sale['Цена за единицу']}") 
        print(f"Общая стоимость: {sale['Общая стоимость']}") 
        print("-" * 40) 
 
if __name__ == "__main__": 
    data = read_sales_data(filecsv) 
    print_sales_data(data)
```

```python
def calculate_total_revenue(obsh_sale_data):
    total_revenue = 0
    for sale in obsh_sale_data:
        total_revenue += float(sale['Общая стоимость'])
    return total_revenue

if __name__ == "__main__":
    obsh_sale_data = read_sales_data(filecsv)
    total_revenue = calculate_total_revenue(obsh_sale_data)
    print(f"Общая выручка: {total_revenue}")
```

```python
def find_best_selling_product(bestsales_data):
    sales_count = defaultdict(int)
    
    for sale in bestsales_data:
        product_name = sale['Название товара']
        sales_count[product_name] += int(sale['Количество продаж'])
    
    best_selling_product = max(sales_count, key=sales_count.get)
    max_sales = sales_count[best_selling_product]
    
    return best_selling_product, max_sales

if __name__ == "__main__":
    bestsales_data = read_sales_data(filecsv)
    best_selling_product, max_sales = find_best_selling_product(bestsales_data)
    print(f"Товар, который был продан наибольшее количество раз: {best_selling_product} (Количество продаж: {max_sales})")
```

```python
def find_highest_revenue_product(maxprofitsales_data):
    revenue_per_product = defaultdict(float)
    
    for sale in maxprofitsales_data:
        product_name = sale['Название товара']
        revenue_per_product[product_name] += float(sale['Общая стоимость'])
    
    highest_revenue_product = max(revenue_per_product, key=revenue_per_product.get)
    highest_revenue = revenue_per_product[highest_revenue_product]
    
    return highest_revenue_product, highest_revenue

if __name__ == "__main__":
    maxprofitsales_data = read_sales_data(filecsv)
    highest_revenue_product, max_revenue = find_highest_revenue_product(maxprofitsales_data)
    print(f"Товар, который принес наибольшую выручку: {highest_revenue_product} (Выручка: {max_revenue})")
```

```python
def print_car_products(cars_data):
    print("Товары категории 'Автомобиль':")
    print("-" * 40)
    for sale in cars_data:
        if sale['Категория товара'] == 'Автомобиль':
            print(f"Название товара: {sale['Название товара']}")
            print(f"Количество продаж: {sale['Количество продаж']}")
            print(f"Цена за единицу: {sale['Цена за единицу']}")
            print(f"Общая стоимость: {sale['Общая стоимость']}")
            print("-" * 40)

if __name__ == "__main__":
    cars_data = read_sales_data(filecsv)
    print_car_products(cars_data)
```

```python
def selection_sort_by_order_number(sortviborka_data):
    for i in range(len(sortviborka_data)):
        min_index = i
        for j in range(i + 1, len(sortviborka_data)):
            if int(sortviborka_data[j]['Номер заказа']) < int(sortviborka_data[min_index]['Номер заказа']):
                min_index = j
        sortviborka_data[i], sortviborka_data[min_index] = sortviborka_data[min_index], sortviborka_data[i]

if __name__ == "__main__":
    sortviborka_data = read_sales_data(filecsv)
    print("До сортировки:")
    print("-" * 40)
    print_sales_data(sortviborka_data)
    
    selection_sort_by_order_number(sortviborka_data)
    
    print("После сортировки:")
    print("-" * 40)
    print_sales_data(sortviborka_data)
```
